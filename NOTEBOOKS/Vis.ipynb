# Paquetes Plots y StatsPlots

```julia
using Plots, StatsPlots
```

```julia
x = 1:10; y = rand(10);
```

```julia
plot(x,y)
```

```julia
x=1:10; y=rand(10,2);
plot(x,y)
```

```julia
plot(x, y, title = "Dos Líneas", label = ["Línea 1" "Línea 2"], lw = 3)
```

```julia
scatter(x,y,title="Diagrama de dispersión")
```

```julia
y = rand(10, 4)
plot(x, y, layout = (4, 1))
```

```julia
p1 = plot(x, y) 
p2 = scatter(x, y) 
p3 = plot(x, y, xlabel = "Título eje x", lw = 3, title = "Subtítulo")
p4 = histogram(x,y) 
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
boxplot(x,y)
```

```julia
z=randn(1000);
histogram(z,color=colormap("Greens",15))
```

```julia
using Distributions
n = 1000
x = rand(Gamma(2), n)
y = -0.5x + randn(n)
marginalhist(x, y, fc = :plasma, bins = 40)
```

```julia
using LaTeXStrings
x=0:0.1:2*π
y=sin.(x).^2
plot(x, y, label=L"$\sin(x)^2$")
```

```julia
x_vals = [0 0 0 ; 2 -3 -4]
y_vals = [0 0 0 ; 4 3 -3.5]

plot(x_vals, y_vals, arrow = true, color = :blue,
     legend = :none, xlims = (-5, 5), ylims = (-5, 5),
     annotations = [(2.2, 4.4, "[2, 4]"),
                    (-3.3, 3.3, "[-3, 3]"),
                    (-4.4, -3.85, "[-4, -3.5]")],
     xticks = -5:1:5, yticks = -5:1:5,
     framestyle = :origin)
```
